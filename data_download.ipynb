### Two Sample T-Test
#### A two-sample t-test compares the means of two independent groups to determine if there is a statistically significant difference between them.

#### H0: The average Deaths / 100 Cases in two different regions considered is equal
#### H1: The average Deaths / 100 Cases in two different regions considered is not equal
#### Assuming alpha = 0.05

```python
st = Statistical_test()
test_stat, p_val = st.two_sample_ttest(df, 'Europe', 'Africa')
```

```python
st.print_values()
```

```python
st.hypothesis_test()
```

### Interested whether a given WHO sample's mean is greater than the entire dataset
#### One Sample One Tailed T -Test
#### alpha = 0.05
#### H0 = The mean of sample < Mean of the whole 
#### H1 = The mean of sample > Mean of whole

```python
st.one_sample_ttest(df, "Recovered / 100 Cases", 'Western Pacific', 'less')
```

```python
st.hypothesis_test()
```

### Proportinality Z-test
#### “Is the percentage of high-recovery countries significantly different between Europe and Africa?”
#### Let proportion be greater than80 to be considered as high-recovery

```python
t, p = st.z_test_proportions(df, 'Europe', 'Africa', 80 )
    
print(t, p)
print(st.hypothesis_test())
```

### Chi-Square Test Hypotheses
#### Testing whether the binary outcome (e.g., High Recovery) is associated with a categorical group (e.g., WHO Region):
#### H0 = The proportion of countries meeting the binary condition is the same across all groups.
#### H1 = The proportion of countries that meet the condition differs significantly across the groups.


```python
t, p, dof, exp = st.chi_square_test_binary(df, 'WHO Region', 'Recovered / 100 Cases', '>', 70)
st.print_values()
```

```python
st.hypothesis_test()
```
